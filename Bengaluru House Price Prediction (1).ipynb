```python
# Convert 'price' from rupees to lakhs
dataset['Price_Lakhs'] = dataset['Price'] / 100000

# Round to 2 decimal places
dataset['Price_Lakhs'] = dataset['Price_Lakhs'].round(2)
print(dataset.head(3))
```

```python
dataset.columns = dataset.columns.str.title()
```

```python
dataset.drop_duplicates()
```

```python
# removing extreme outliers in price column
Q1 = dataset['Price'].quantile(0.25)
Q3 = dataset['Price'].quantile(0.75)
IQR = Q3 - Q1
df = dataset[(dataset['Price'] >= Q1 - 1.5 * IQR) & (dataset['Price'] <= Q3 + 1.5 * IQR)]
```

```python
# Ensure 'price' column exists (lowercase as per original dataset)
dataset['Price_Lakhs'] = dataset['Price'] / 100000
dataset['Price_Lakhs'] = dataset['Price_Lakhs'].round(2)

sns.histplot(dataset['Price_Lakhs'])
```

```python
# Price by Location

plt.figure(figsize=(10,5))
sns.boxplot(x='Location', y='Price_Lakhs', data=dataset)
plt.title('Price by Location')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 Locations by Average Price

top_locations = dataset.groupby('Location')['Price_Lakhs'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.xticks(rotation=45)
plt.show()
```

```python
# Price by bedrooms

plt.figure(figsize=(10,5))
sns.boxenplot(x='Bhk', y='Price_Lakhs', data=dataset)
plt.title('Price by bhk')
plt.show()
```

```python
# Price by Furnishing

plt.figure(figsize=(10,5))
sns.stripplot(x='Furnishing', y='Price_Lakhs', data=dataset)
plt.title('Price by Furnishing')
plt.show()
```
